# join method

```python
data1 = pd.DataFrame({"key1": [1,2,3,4,5],
                      "key2": [6,7,8,9,10],
                      "key3": [11,12,13,14,15]}, index = ['a','b','c','d','e'])
```

```python
data2 = pd.DataFrame({"key11": [1,20,3,40,5],
                      "key22": [6,70,8,9,10],
                      "key33": [131,120,139,14,150]}, index = ['a','b','f','g','h'])
```

```python
data1
```

```python
data2
```

```python
data1.join(data2)
```

```python
data1.join(data2, how = "right")
```

```python
data1.join(data2, how = "outer")
```

```python
data1.join(data2, how = "cross")
```
